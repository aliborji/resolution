```
def plot_data(all_data, y_range):

  plt.figure(figsize=(10,2.5), dpi= 80)
  plt.subplot(141)

  titles = ['12 px', '16 px', '20 px', '24 px'] 


  avgs, stds = [], []

  for idx, data in enumerate(all_data):

    means = np.array([(x[-1],np.max(y)) for x,y in data])#, axis=0)

    ax = plt.subplot(1,4,idx+1)
    x = np.arange(len(means))


    tmp = [u[-1] for u,_ in data]
    ax.scatter(x, tmp, marker='o', color='k')

    for t in x:
      tmp = data[t][1]
      ax.scatter([t]*len(tmp), tmp, marker='.', color='r')


    # ax.bar(x-0.2, means[:,0], width=0.3, color='b', align='center')
    # ax.bar(x, means[:,1], width=0.3, color='g', align='center')

    ax.set_title(f'{titles[idx]},   num exps={sum([len(y) for _,y in data])}')
    if idx:
      ax.set_yticklabels([])

    ax.set_xticks(x)
    ax.set_xticklabels(x+1)  

    avgs.append(np.mean(means, axis=0))
    stds.append(np.std(means, axis=0))

    plt.ylim(y_range)
    plt.grid(linestyle='--', axis='x')

  plt.tight_layout()
  plt.show()

  return avgs, stds
```

```
grand_avgs, grand_stds = [], []
all_data = [data_mnist, data_fmnist, data_cifar]
y_ranges = [y_range_mnist, y_range_fmnist, y_range_cifar]

for data, yr in zip(all_data, y_ranges):
  a, b = plot_data(data, yr)
  grand_avgs.append(a)
  grand_stds.append(b)

```
